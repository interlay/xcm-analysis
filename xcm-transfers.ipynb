```python

import json
import pandas as pd

# Opening JSON file
f = open('data/xcm-moonbeam-interlay.json')
  
# returns JSON object as 
# a dictionary
moonbeam_interlay = json.load(f)
moonbeam_interlay = pd.DataFrame(moonbeam_interlay).fillna(0)
moonbeam_interlay = moonbeam_interlay.astype({"blockNumber": "int32", "blockNumberDest": "int32"})


filtered_moonbeam_data = moonbeam_interlay.loc[(moonbeam_interlay["rawAsset"] == '{"Token":"IBTC"}')
                      & (moonbeam_interlay["blockNumberDest"] > 975127)
                      & (moonbeam_interlay["blockNumber"] > 1638215)]

print(f"Number of xcm transactions after the exploit: {len(filtered_moonbeam_data)}")
filtered_moonbeam_data.loc[:, ["amountSent", "amountReceived"]].sum()
```
